Requirements

1. Input the csv file
2. Add a row number to the data set
3. Find the latest row (largest row number) to capture the individuals correct salary information
4. Find each team member's annual salary
5. Find each team member's maximum tax band based on their annual salary
    - 20% rate
    - 40% rate
    - 45% rate 
6. Work out how much tax an individual paid for each of the % bands. Call these fields:
    - 20% tax rate paid
    - 40% tax rate paid
    - 45% tax rate paid
7. Total the tax paid across all three % bands. Call this field 'Total Tax Paid' 
8. Output the data

```python
import pandas as pd
```

1. Input the csv file
2. Add a row number to the data set

```python
income_df = pd.read_csv('PD 2024 Wk 6 Input.csv')
```

```python
income_df
```

3. Find the latest row (largest row number) to capture the individuals correct salary information
4. Find each team member's annual salary

```python
income_df['Salary'] = income_df.iloc[:, 1:13].sum(axis=1)
```

```python
income_df
```

5. Find each team member's maximum tax band based on their annual salary
    - 20% rate
    - 40% rate
    - 45% rate 

```python
def tax_band(sal):
    if sal > 12570 and sal <= 50270:
        return '20% Rate'
    elif sal > 50270 and sal <= 125140:
        return '40% Rate'
    else:
        return '45% Rate'
```

```python
income_df['Max Tax Rate'] = income_df['Salary'].apply(tax_band)
```

```python
income_df
```

6. Work out how much tax an individual paid for each of the % bands. Call these fields:
    - 20% tax rate paid
    - 40% tax rate paid
    - 45% tax rate paid

```python
def tax_20_pct_paid(max_tax,sal):
    if max_tax == '20% Rate':
        return round(((sal - 12570) * 0.2),2)
    else:
        return round(((50270 - 12570)*0.2),2)
```

```python
income_df['20% Tax Rate Paid'] = income_df.apply(lambda x: tax_20_pct_paid(x['Max Tax Rate'], x['Salary']), axis=1 ) 
```

```python
income_df
```

```python
def tax_40_pct_paid(max_tax,sal):
    if max_tax == '40% Rate':
        return round(((sal - 50270) * 0.4),2)
    elif max_tax == '45% Rate':
        return round(((125140 - 50270)*0.4),2)
    else:
        return 0
```

```python
income_df['40% Tax Rate Paid'] = income_df.apply(lambda x: tax_40_pct_paid(x['Max Tax Rate'], x['Salary']), axis=1 ) 
```

```python
income_df
```

```python
def tax_45_pct_paid(max_tax,sal):
    if max_tax == '45% Rate':
        return round(((sal - 125140) * 0.45),2)
    else:
        return 0
```

```python
income_df['45% Tax Rate Paid'] = income_df.apply(lambda x: tax_45_pct_paid(x['Max Tax Rate'], x['Salary']), axis=1 ) 
```

```python
income_df
```

7. Total the tax paid across all three % bands. Call this field 'Total Tax Paid' 

```python
income_df['Total Tax Paid'] = income_df.iloc[:, 15:18].sum(axis=1)
```

```python
income_df
```

8. Output the data

```python
income_df = income_df[['StaffID','Salary','Max Tax Rate','20% Tax Rate Paid','40% Tax Rate Paid','45% Tax Rate Paid','Total Tax Paid']]
```

```python
income_df
```
