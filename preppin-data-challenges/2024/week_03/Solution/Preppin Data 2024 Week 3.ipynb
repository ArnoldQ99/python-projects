```python
import pandas as pd
```

1. Input the outputs from 2024 Week 1 challenge.
2. Input the new targets Excel sheet (Q1 - 4).

```python
flow_card_df = pd.read_csv('Preppin Data Inputs/Week 3/PD 2024 Wk 1 Output Flow Card.csv')
```

```python
non_flow_card_df = pd.read_csv('Preppin Data Inputs/Week 3/PD 2024 Wk 1 Output Non-Flow Card.csv')
```

```python
flow_card_df
```

```python
non_flow_card_df
```

```python
sales_df = pd.concat([flow_card_df,non_flow_card_df])
```

```python
sales_df
```

```python
sheets = pd.read_excel('Preppin Data Inputs/Week 3/PD 2024 Wk 3 Input.xlsx',sheet_name=None)
```

```python
targets_df = pd.concat(sheets, ignore_index=True)
```

```python
targets_df
```

3. Correct the Classes being incorrect as per last week: Economy to First, First Class to Economy, Business Class to Premium, Premium Economy to Business.

```python
sales_df['Class'] = sales_df['Class'].replace({'Economy':'First Class','First Class':'Economy','Business Class':'Premium Economy','Premium Economy':'Business Class'})
```

```python
sales_df
```

4. Find the First Letter from each word in the Class to help with joining the Targets data to Sales data

```python
sales_df['Class'] = sales_df['Class'].apply(lambda x: ''.join(word[0] for word in x.split(' ')))
```

```python
sales_df
```

5. Change the date to a month number

```python
sales_df['Date'] = pd.to_datetime(sales_df['Date'],dayfirst=True).dt.month
```

```python
sales_df
```

6. Total up the sales at the level of: Class & Month

```python
agg_df = sales_df.groupby(['Class','Date'],as_index=False)['Price'].sum()
```

```python
agg_df
```

7. Join the Targets data on to the Sales data (note - you should have 48 rows of data after the join)

```python
df = pd.merge(left=agg_df, right=targets_df, how='inner', left_on=['Date','Class'], right_on=['Month','Class'])
```

```python
df
```

8. Calculate the difference between the Sales and Target values per Class and Month

```python
df['Difference to Target'] = df['Price'] - df['Target']
```

```python
df
```

```python
df = df.drop(columns='Month')
```

```python
df
```

9. Output the data

```python
df.to_csv('Preppin Data Outputs/pd2024wk3_output.csv')
```
