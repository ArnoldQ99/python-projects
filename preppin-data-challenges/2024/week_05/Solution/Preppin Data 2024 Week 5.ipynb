```python
import pandas as pd
```

1. Input the data

```python
sales_df = pd.read_csv('Prep Air Ticket Sales.csv')
```

```python
sales_df
```

```python
flights_df = pd.read_csv('Prep Air 2024 Flights.csv')
```

```python
flights_df
```

```python
customers_df = pd.read_csv('Prep Air Customers.csv')
```

```python
customers_df
```

2. For the first output:
    - Create a dataset that gives all the customer details for booked flights in 2024. Make sure the output also includes details on the  flights origin and destination
    - When outputting the data, create an excel file with a new sheet for each output (so 1 file for all outputs this week!)

```python
output1_df = flights_df.merge(right=sales_df, how='left', on=['Flight Number','Date'])
```

```python
output1_df
```

```python
output1_df = output1_df.merge(right=customers_df, how='inner', on='Customer ID')
```

```python
output1_df
```

3. For the second output:
    - Create a dataset that allows Prep Air to identify which flights have not yet been booked in 2024
    - Add a datestamp field to this dataset for today's date (31/01/2024) so that Prep Air know the unbooked flights as of the day the workflow is run
    - When outputting the table to a new sheet in the Excel Workbook, choose the option "Append to Table" under Write Options. This means that if the workflow is run on a different day, the results will add additional rows to the dataset, rather than overwriting the previous run's data

```python
output2_df = sales_df.merge(right=flights_df, how='outer', on=['Date','Flight Number'])
```

```python
output2_df 
```

```python
output2_df = output2_df[output2_df['Customer ID'].isnull()]
```

```python
output2_df = output2_df.copy()
```

```python
output2_df['Flights Unbooked As Of'] = '31/01/2024'
```

```python
output2_df['Flights Unbooked As Of'] = pd.to_datetime(output2_df['Flights Unbooked As Of'], dayfirst=True)
```

```python
output2_df
```

4. For the third output:
    - Create a dataset that shows which customers have yet to book a flight with Prep Air in 2024
    - Create a field which will allow Prep Air to see how many days it has been since the customer last flew (compared to 31/01/2024)
    - Categorise customers into the following groups:
        - Recent fliers - flown within the last 3 months
        - Taking a break - 3-6 months since last flight
        - Been away a while - 6-9 months since last flight
        - Lapsed Customers - over 9 months since last flight
    - Output the data to a new sheet in the Excel Workbook

```python
output3_df = sales_df.merge(right=flights_df, how='inner' , on=['Date','Flight Number'])
```

```python
output3_df = output3_df.merge(right=customers_df, how='right', left_on='Customer ID',right_on='Customer Id', right_index=True).reset_index()
```

```python
output3_df
```

```python
output3_df = output3_df[output3_df['Customer ID_x'].isnull()]
```

```python
output3_df = output3_df.drop(output3_df[output3_df['Customer ID'] == 0 ].index)
```

```python
output3_df = output3_df[['Customer ID', 'Last Date Flown', 'first_name','last_name','email','gender']]
```

```python
output3_df
```

```python
output3_df = output3_df.copy()
```

```python
output3_df['Last Date Flown'] = pd.to_datetime(output3_df['Last Date Flown'],yearfirst=True)
```

```python
output3_df['Days since last flown'] = (pd.to_datetime('2024-01-31', yearfirst=True) - output3_df['Last Date Flown']).dt.days
```

```python
output3_df
```

```python
def cats(x):
    if x < 90:
        return 'Recent fliers'
    elif x >= 90 and x < 180:
        return 'Takinng a break 3-6 months since last flight'
    elif x >= 180 and x < 270:
        return 'Been away a while - 6-9 months since last flight'
    else:
        return 'Lapsed Customers - over 9 months since last flight'
```

```python
output3_df['Customer Category'] = output3_df['Days since last flown'].apply(cats)
```

```python
output3_df
```

```python
with pd.ExcelWriter('week_05_output.xlsx') as writer:
    output1_df.to_excel(writer, sheet_name='First Output', index=False)
    output2_df.to_excel(writer, sheet_name='Second Output', index=False)
    output3_df.to_excel(writer, sheet_name='Third Output', index=False)
```
