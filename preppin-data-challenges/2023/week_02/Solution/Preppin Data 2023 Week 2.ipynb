```python
import pandas as pd
```

```python
# 1. Input the data

swift_codes = pd.read_csv('Preppin Data Inputs/Swift Codes.csv')

transactions = pd.read_csv('Preppin Data Inputs/Transactions.csv')
```

```python
swift_codes
```

```python
transactions
```

```python
# 2. In the Transactions table, there is a Sort Code field which contains dashes. We need to remove these so just have a 6 digit string

transactions['Sort Code'] = transactions['Sort Code'].str.replace('-', '')
```

```python
transactions
```

```python
# 3. Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and Check Digits.

merge_data = transactions.merge(right=swift_codes ,how='left', left_on='Bank', right_on='Bank')
```

```python
# 4. Add a field for the Country Code

country_code = 'GB'

merge_data.insert(loc=0 ,column='Country code', value=country_code)
```

```python
merge_data
```

```python
# 5. Create the IBAN as above

merge_data['IBAN'] = (merge_data['Country code'].astype(str) + 
     merge_data['Check Digits'].astype(str) + 
     merge_data['SWIFT code'].astype(str) +
     merge_data['Sort Code'].astype(str) +
     merge_data['Account Number'].astype(str)
    )
```

```python
merge_data
```

```python
# 6. Remove unnecessary fields

merge_data = merge_data.drop(['Country code', 'Account Number', 'Sort Code', 'Bank', 'SWIFT code', 'Check Digits'], axis=1)
```
